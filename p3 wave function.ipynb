```python

def get_super(x): 
    normal = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+-=()"
    super_s = "ᴬᴮᶜᴰᴱᶠᴳᴴᴵᴶᴷᴸᴹᴺᴼᴾQᴿˢᵀᵁⱽᵂˣʸᶻᵃᵇᶜᵈᵉᶠᵍʰᶦʲᵏˡᵐⁿᵒᵖ۹ʳˢᵗᵘᵛʷˣʸᶻ⁰¹²³⁴⁵⁶⁷⁸⁹⁺⁻⁼⁽⁾"
    res = x.maketrans(''.join(normal), ''.join(super_s)) 
    return x.translate(res) 
def get_sub(x): 
    normal = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+-=()"
    sub_s = "ₐ₈CDₑբGₕᵢⱼₖₗₘₙₒₚQᵣₛₜᵤᵥwₓᵧZₐ♭꜀ᑯₑբ₉ₕᵢⱼₖₗₘₙₒₚ૧ᵣₛₜᵤᵥwₓᵧ₂₀₁₂₃₄₅₆₇₈₉₊₋₌₍₎"
    res = x.maketrans(''.join(normal), ''.join(sub_s)) 
    return x.translate(res) 
from sympy.physics.wigner import wigner_6j
import numpy as np
import pandas as pd
from numpy import nan as nan
D=[]
s1=[]
l=1
spin=1/2


orbital1='p'
orbital11=2
orbital2='p'
orbital22=1

orbital3='p3' # Excel sheet name


# Parent term
L_d=[0,2,1]
S_d=[0,0,1]
L_dd=['S','D','P']
S_dd=[1,1,3]




#Final term

L=[1,2,0]
S=[0.5,0.5,1.5]
L1=['P','D','S']
S1=[2,2,4]



for j in range(len(L)):
    x2=L[j]
    z2=S[j]
    z22=S1[j]
    x22=L1[j]



    for ku in range(len(L_d)):
        B11=[]
        B1=[]
        x1=L_d[ku]
        z1=S_d[ku]
        x11=L_dd[ku]
        z11=S_dd[ku]
        for i in range(len(L_d)): #3
            x3=L_d[i]
            z3=S_d[i]
            delta=0
            if x3==x1 and z3==z1:
               delta =1
            else:
               delta = 0
            a=2*np.sqrt(((2*x1)+1)*((2*z1)+1)*((2*x3)+1)*((2*z3)+1))
            b=wigner_6j(spin,spin,z3,spin,z2,z1)*wigner_6j(l,l,x3,x2,l,x1)
            c=delta+(a*b)
            B1.append(c)
     
        #if len(B1)!=0:
        Nt=0
        for k1 in range(len(B1)):
            Nt=(B1[k1])**2+Nt
        for n in range(len(B1)):
            s=float(B1[n]/(Nt**0.5))
            s=float(format(s, ".6f"))
            B11.append(s)

        
        Norm=0
        for k1 in range(len(B11)):
            Norm=(B11[k1])**2+Norm
            Norm=float(format(Norm, ".6f"))
        Wave_eqn='|{}{}{}{};({}{}){}{}>='.format(str(f'{orbital1}'),get_super(f'{orbital11}'),str(f'{orbital2}'),get_super(f'{orbital22}'),get_super(f'{z11}'),str(f'{x11}'),get_super(f'{z22}'),str(f'{x22}'))
        for i in range(len(B11)):
            a=B11[i]
            x3=L_d[i]
            z3=S_d[i]
            x33=L_dd[i]
            z33=S_dd[i]
            if a !=0:
               Wave_eqn=Wave_eqn+'({})|{}{}{}{};({}{}){}{}>'.format(str(f'{a}'),str(f'{orbital1}'),get_super(f'{orbital11}'),str(f'{orbital2}'),get_super(f'{orbital22}'),get_super(f'{z33}'),str(f'{x33}'),get_super(f'{z22}'),str(f'{x22}'))+'+'
        Wave_eqn=Wave_eqn[:-1]
        s1.append(Wave_eqn)
        D.append(Norm)
#d7=list(range(1,(len(L_d)*len(L))+1))
p=pd.DataFrame(list(zip(s1,D)),columns=["Wave Function","Normalization"])
a1=p.dropna(axis=0)
a1.to_excel(f"{orbital3}.xlsx",index=False)
p
```
